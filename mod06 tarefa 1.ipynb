```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib notebook
```

```python
sinasc = pd.read_csv('base_sinac.csv')
sinasc.head()
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
# enumerate...?
# flatten...?
colunas = ['APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE']
sin = sinasc[colunas]

contar = {}
for col in sin.columns:
  contar[col] = sin[col].value_counts()

fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 8))
for i, ax in enumerate(axes.flatten()):
  if i < len(colunas):
    contar[colunas[i]].plot.bar(ax=ax)
    ax.set_title(colunas[i])
  else:
    ax.axis('off')
plt.tight_layout()
plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
fig, ax = plt.subplots()
sns.barplot(x='APGAR5', y='CONSULTAS', data=sinasc)
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
colunas = ['SEXO', 'GESTACAO', 'ESTCIVMAE', 'CONSULTAS']
sin = sinasc[colunas]

contar = {}
for col in sin.columns:
  contar[col] = sin[col].value_counts()

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 8))
for i, ax in enumerate(axes.flatten()):
  if i < len(colunas):
    contar[colunas[i]].plot.bar(ax=ax)
    ax.set_title(colunas[i])
  else:
    ax.axis('off')
plt.tight_layout()
plt.show()
```
